```python
import pandas as pd
```

```python
df = pd.read_csv('gate_cutoff.csv')
df.head()
```

```python
df = df.drop(columns = ['Unnamed: 0'])
df.rename(columns = {'Sr No':'srNo','Paper Code':'paperCode','Paper Name':'paperName','Year':'year','GEN':'gen','Code':'code','OBC':'obc','SC/ST':'sc/st'},inplace = True)
df.head()
```

```python
df['paperName'].unique()
```

```python
filter = (df['paperName'] == 'Computer Science and Information Technology')
df1 = df[filter]
df1
```

```python
k = df[['year','paperName','gen','obc','sc/st']].groupby(['year','paperName'])
```

```python
k = k.apply(lambda x:x)
```

```python
k
```

```python
k = k.sort_values(by = ['year','gen'], ascending = [False,False])
k.style.background_gradient(cmap = 'Reds',subset=['gen','obc','sc/st'])
```

```python
k = df[['paperName','gen','obc','sc/st']].groupby('paperName')['gen','obc','sc/st'].mean()
k = k.sort_values(by = ['gen'], ascending = [False])
k.style.background_gradient(cmap = 'Reds',subset=['gen','obc','sc/st'])
```
